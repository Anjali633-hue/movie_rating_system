**Step-4** What are the most common movie genres?

```python
genre_counts = data['Genre'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette="viridis")
plt.xticks(rotation=45)
plt.xlabel("Genres")
plt.ylabel("Number of Movies")
plt.title("Most Common Movie Genres")
plt.show()
```

**Step-5** Which studios produce the highest number of movies?

```python
studio_counts = data['Lead Studio'].value_counts().head(10)  # Top 10 studios
plt.figure(figsize=(10,5))
sns.barplot(x=studio_counts.index, y=studio_counts.values, palette="magma")
plt.xticks(rotation=45)
plt.xlabel("Studios")
plt.ylabel("Number of Movies")
plt.title("Top 10 Studios Producing the Most Movies")
plt.show()
```

**Step-6** What is the distribution of movie ratings?

```python
plt.figure(figsize=(8,5))
sns.histplot(data['rating'], bins=10, kde=True, color="blue")
plt.xlabel("Ratings")
plt.ylabel("Frequency")
plt.title("Distribution of Movie Ratings")
plt.show()
```

**Step-7** Which movies have the highest profitability?

```python
data['Profitability'].value_counts()
plt.figure(figsize=(6,5))
sns.countplot(x='rating',order=data['rating'].value_counts().index[0:10],data=data)
plt.xlabel("rating")
plt.ylabel("count")
plt.title("Distribution of Movie Ratings")
plt.show()
```

**Step-8** highest rating of two column

```python
plt.figure(figsize=(6,5))
sns.barplot(y=data['Film'].head(20),x=data['rating'].head(20),data=data)
plt.xlabel('rating')
plt.ylabel('Film')
plt.title("In top 20 movies which movie gains highest rating")
plt.show()
```
